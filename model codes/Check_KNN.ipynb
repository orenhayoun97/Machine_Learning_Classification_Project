```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
def LoadData(X_filename,Y_filename):
    '''
    Input: Get 2 filenames, 1 for x_data and 1 for y_data
    Output: Two numpy array of X and Y data
    '''
    X = pd.read_csv(X_filename,delimiter=';')
    Y = pd.read_csv(Y_filename,delimiter=';')
    feature_names = X.columns
    X = X.to_numpy()
    Y = Y.to_numpy()
    return X,Y
    
```

```python
def GetFeatureNames(X_filename):
    X = pd.read_csv(X_filename,delimiter=';')
    feature_names = X.columns
    return feature_names
```

```python
def SplitTrainTestRandom(X,Y):
    '''
    ***COMPLETED - NO CHANGES ARE REQUIRED***
    
    Input: feature matrix X, labels vector Y
    Output: X_train, Y_train - train data
            X_test, Y_test - test data
    '''
    
    # permute the ordering of the examples
    # random inds every running 
    ind = np.random.permutation(len(Y))
    
    # choose the size of the training data
    Ntrain = int(0.8 * len(Y))
    # split the data into train and test datasets
    
    X_train = X[ind[:Ntrain]]
    Y_train = Y[ind[:Ntrain]]
    X_test  = X[ind[Ntrain:]]
    Y_test  = Y[ind[Ntrain:]]
    
    return X_train, Y_train, X_test, Y_test
```

## cross-validation

```python
X,Y = LoadData("X_data.csv","Y_data.csv")
running_num = 20
acc_array = np.array([])
K_array = np.array([35,50,100,200,500,1000])
#accuracy_every_k = np.zeros((Kmax,running_num)) # create a matrix Kmax rows and running_num columns
# repeat for all considered values of K
for k in K_array:
    acc = np.array([])
    for i in range(running_num):
        X_train, Y_train, X_test, Y_test = SplitTrainTestRandom(X,Y)
        clf = KNeighborsClassifier(n_neighbors=k).fit(X_train,Y_train)
        acc = np.append(acc,clf.score(X_test,Y_test))
    acc_array = np.append(acc_array,acc.mean())

plt.figure(figsize=(14,5))
plt.plot(K_array,acc_array,'.-')
plt.plot([K_array[acc_array.argmax()],K_array[acc_array.argmax()]],[0,acc_array.max()],'-o')
plt.legend(['acc(K_array)',f'acc({K_array[acc_array.argmax()]})={np.round(acc_array.max(),3)}'])
plt.xlabel('K')
plt.ylabel('Accuracy')
plt.title('KNN estimated accuracy for 20 running and calculate mean for each K');
```

## Optimum neighbors

```python
X,Y = LoadData("X_data.csv","Y_data.csv")
running_num = 50
acc = np.array([])
for i in range(running_num):
    X_train, Y_train, X_test, Y_test = SplitTrainTestRandom(X,Y)
    clf = KNeighborsClassifier(n_neighbors=29).fit(X_train,Y_train)
    acc = np.append(acc,clf.score(X_test,Y_test))
acc.mean()
```

```python
from sklearn.inspection import permutation_importance
result = permutation_importance(clf, X_test, Y_test, n_repeats=40, random_state=42)
feature_names=GetFeatureNames("X_data.csv")
importances = result.importances_mean

plt.figure(figsize=(20,5))
plt.bar(range(len(importances)), importances) 
plt.xticks(ticks=range(len(feature_names)),labels=feature_names)
plt.show()
```
